```python
import pikepdf
from plyer import notification
from tqdm import tqdm
p=input("password list: ")
ppdf=input("protected pdf: ")
# load password list
passwords = [line.strip() for line in open(p)]

# iterate over passwords
for password in tqdm(passwords, "Decrypting PDF"):
    try:
        # open PDF file
        with pikepdf.open(ppdf, password=password) as pdf:
            # Password decrypted successfully, break out of the loop
            notification_title = 'GREETINGS FROM JAVATPOINT!'  
            notification_message = 'Thank you for reading. Have a Good Day.'  
            
  
            notification.notify(  
                title =notification_title,
                message = password,  
                app_icon =None ,  
                timeout = 20,  
                toast = True  
                )  
            break
        
    except pikepdf._qpdf.PasswordError as e:
        # wrong password, just continue in the loop
        continue

        
```
